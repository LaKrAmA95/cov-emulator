```python

import numpy as np
import matplotlib.pyplot as plt
```

```python

d = np.arange(100)/100
plt.plot(d,np.exp(-d/ 5), label='p_dist')
plt.xlabel("Distance")
plt.ylabel("Transmission Probability")
plt.ylim(0,1)
```

```python
t = np.arange(60)
p=((np.tanh(t - 5) + 0.2) * (np.tanh(-t + 8) + 0.5) + 1.19987) / 2.6683940
plt.plot(t,p, label='p_dist')
plt.xlabel("Infected duration")
plt.ylabel("Transmission Probability")
plt.ylim(0,1)
```

```python

c = np.arange(1,11)
c2 = c / 20 + 0.5
p = c2 + (1 - c2) ** 2
plt.plot(c,p, label='p_dist')
plt.xlabel("Number of contacts")
plt.ylabel("Transmission Probability")
plt.ylim(0,1)
```

```python
x,y =10,10
r1 = 10
r2 = 1

plt.figure()
ax = plt.gca()
c = plt.Circle((x,y),r1, facecolor=[0,0,0,0], edgecolor='k')
ax.add_patch(c)
ax.set_xlim(-r1-1+x,r1+1+x)
ax.set_ylim(-r1-1+y,r1+1+y)
ax.set_aspect('equal', adjustable='box')

for _r in range((r1-r2)//(2*r2)):
    _r = r1 - r2 -2*r2*_r
    
    theta = np.arcsin(r2/_r)
    print(_r, theta, np.pi//theta, np.pi%theta)
    for _theta in range(int(np.pi*1000)//int(theta*1000)):
        
        _theta = theta*2*_theta
        _x = _r*np.cos(_theta)
        _y = _r*np.sin(_theta)
        c = plt.Circle((_x+x,_y+y),r2, facecolor=[0,0,0,0], edgecolor='k')
        ax.add_patch(c)
if _r > r2:
    c = plt.Circle((x,y),r2, facecolor=[0,0,0,0], edgecolor='k')
    ax.add_patch(c)
plt.show()
```
